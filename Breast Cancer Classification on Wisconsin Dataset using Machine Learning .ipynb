## Create and Train the Classifier

**Reference Links :**

Sklearn SVM : https://scikit-learn.org/stable/modules/classes.html#module-sklearn.svm

Sklearn SVC : https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC

```python
from sklearn.svm import SVC
```

```python
svc_classifier = SVC()
```

```python
svc_classifier.fit(X_train, y_train)
```

## Evaluate the Model/Classifier

**Reference Links :**

Sklearn Metrics : https://scikit-learn.org/stable/modules/classes.html#module-sklearn.metrics

Sklearn Confusion Matrix : https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_predicted = svc_classifier.predict(X_test)
```

```python
y_predicted
```

```python
# Confusion Matrix

conf_mat = confusion_matrix(y_true=y_test,y_pred=y_predicted)
```

```python
conf_mat
```

```python
# Plot the heat map of the Confusion Matrix for better understanding

sns.heatmap(data=conf_mat,annot=True)
```

```python
# Classification Report 

print(classification_report(y_true=y_test, y_pred=y_predicted))
```

***Above Classification Report is bad. It indicates that the Model is not classifying the Class 0***

## Improve the Model

### Feature Scaling the data by Unity Normalization i.e. Bring all the values between 0 and 1. It is a common practice to improve the model's efficacy

**Reference Links :** 

Sklearn Processing : https://scikit-learn.org/stable/modules/preprocessing.html

Sklearn MinMaxScaler : https://scikit-learn.org/stable/modules/preprocessing.html#scaling-features-to-a-range

Sklearn MinManScaler : https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html



Rescaling the Data for Machine Learning : https://machinelearningmastery.com/rescaling-data-for-machine-learning-in-python-with-scikit-learn/

Prepare Data for Machine Learning : https://machinelearningmastery.com/prepare-data-machine-learning-python-scikit-learn/

Standardize or Normalize : https://medium.com/@rrfd/standardize-or-normalize-examples-in-python-e3f174b65dfc

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
print(X_train_scaled)
```

```python
X_test_scaled = scaler.fit_transform(X_test)
```

```python
print(X_test_scaled)
```

## Train the Classifier with the scaled values of X and Evaluate

```python
svc_classifier2 = SVC()
```

```python
svc_classifier2.fit(X=X_train_scaled,y = y_train)
```

```python
y_predicted_scaled = svc_classifier2.predict(X_test_scaled)
```

```python
y_predicted_scaled
```

```python
# Confusion Matrix

conf_mat2 = confusion_matrix(y_true=y_test,y_pred=y_predicted_scaled)
```

```python
print(conf_mat2)
```

```python
# Plot the Heat Map for Visualization of the Confusion Matrix

sns.heatmap(data=conf_mat2, annot=True)
```

```python
# Classification Report

print (classification_report(y_true=y_test, y_pred=y_predicted_scaled))
```

***The above Classification Report looks Great ! We have about 98 % accuracy of prediction. So, we draw the inference that feature scaling improved the Model's Efficacy in this case.***

## Explore the Hyper Tuning Parameters

Although we are getting 98% accuracy with the default tuning parameters as below -

SVC(C=1.0, cache_size=200, class_weight=None, coef0=0.0,
  decision_function_shape='ovr', degree=3, gamma='auto_deprecated',
  kernel='rbf', max_iter=-1, probability=False, random_state=None,
  shrinking=True, tol=0.001, verbose=False)
  
We can still use GridSearch to find values of C and Gamma; which **MAY BE** even better; thereby improving the efficacy of the model further

**Reference :**

Parameter Estimation Using Grid Search with Cross Validation : https://scikit-learn.org/stable/auto_examples/model_selection/plot_grid_search_digits.html

Tuning Hyper Parameters of an Estimator : https://scikit-learn.org/stable/modules/grid_search.html

Sklearn GridSearchCV : https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html


```python
from sklearn.model_selection import GridSearchCV
```

```python
tuned_parameters = [{'kernel': ['rbf'], 'gamma': [1, 0.1, 0.01, 0.001,0.0001],
                     'C': [0.1, 1, 10, 100, 1000]},
                    {'kernel': ['linear'], 'C': [1, 10, 100, 1000]}]
```

```python
grid_tuned_svc_classifier = GridSearchCV(estimator=SVC(),param_grid=tuned_parameters,refit=True,cv=5,verbose=4)
```

```python
grid_tuned_svc_classifier.fit(X=X_train_scaled,y=y_train)
```

```python
grid_tuned_svc_classifier.best_params_
```

```python
grid_tuned_svc_classifier.best_estimator_
```

```python
grid_y_predictions = grid_tuned_svc_classifier.predict(X=X_test_scaled)
```

```python
# Confusion Matrix

grid_conf_mat = confusion_matrix(y_true=y_test, y_pred=grid_y_predictions)
```

```python
print(grid_conf_mat)
```

```python
# Plot the Heat Map for Visualization of the Confusion Matrix

sns.heatmap(grid_conf_mat, annot=True)
```

```python
# Classification Report 

print(classification_report(y_test,grid_y_predictions))
```

#### GridSearch's Accuracy is worse as compared to the default model's accuracy. Additional, as seen from the heat map; the GridSeach Tuned Classifier give False Negative which is a risky prediction in the Cancer Detection Use Case. 

#### In this case, a False Negative means that the Patient has Cancer but the Classifier predicted that the Patient does not have Cancer.
